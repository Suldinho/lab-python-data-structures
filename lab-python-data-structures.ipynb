````python
#9 Update the inventory by subtracting 1 from the quantity of each product. Modify the inventory dictionary accordingly. 
# Define the list of products
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Define the initial inventory with quantities for each product
inventory = {
    "t-shirt": 10,
    "mug": 10,
    "hat": 10,
    "book": 10,
    "keychain": 10
}

# Create an empty set for customer orders
customer_orders = set()

# Function to take customer order input
def take_customer_orders():
    print("Please enter the names of three products you want to order from the following list:")
    print(products)
    
    for i in range(3):
        while True:
            product = input(f"Enter product {i + 1}: ").strip().lower()
            if product in products:
                customer_orders.add(product)
                break
            else:
                print("Invalid product name. Please choose from the list above.")
                
    print("Customer orders have been updated.")
    print("Customer orders:", customer_orders)

# Function to update the inventory based on customer orders
def update_inventory():
    for product in customer_orders:
        if inventory[product] > 0:
            inventory[product] -= 1

# Call the function to take customer orders
take_customer_orders()

# Print the products in the customer_orders set
print("Products in the customer orders set:")
for product in customer_orders:
    print(product)

# Update the inventory
update_inventory()

# Print the updated inventory
print("\nUpdated Inventory:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")

# Calculate order statistics
total_products_ordered = len(customer_orders)
percentage_products_ordered = (total_products_ordered / len(products)) * 100

# Store statistics in a tuple
order_status = (total_products_ordered, percentage_products_ordered)

# Print order statistics using the specified format
print("\nOrder Statistics:")
print(f"Total Products Ordered: {order_status[0]}")
print(f"Percentage of Products Ordered: {order_status[1]:.2f}%")
````
